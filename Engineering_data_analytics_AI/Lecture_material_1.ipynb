# Loops
Both for and while are control flow statements in programming languages that allow you to create loops, which are used for repeating a block of code multiple times.\
The for loop is typically used when you know in advance how many times you want to iterate, and you have a sequence (list, tuple, string, etc.) to iterate over.\

The while loop is used when you don't know in advance how many times the loop needs to run. It continues iterating as long as the specified condition is True.\

Use a for loop when you know the number of iterations in advance and have a sequence to iterate over. Use a while loop when the number of iterations is not known in advance and is determined by a condition


```python
# For loop
foods=['bacon','tuna','ham','sausages','beef','eggs','chicken']
for food in foods:
	print(food)
```

```python
# write good example of for loop
for x in range(1,11):
    print(x) 
```

```python
for x in range(1,11,2):
    print(x)
```

```python
# While loop
i=1
while i<=20:
    print(i)
    i+=2
```

```python
# for loop with if else
for x in range(1,11):
    if x%2==0:
        print(x,"is even")
    else:
        print(x,"is odd")
```

```python
# for nested loops
colors=["red","green","blue","yellow","orange","purple"]
items=["shirt","hat","socks","pants","jacket","shorts"]
for color in colors:
    for item in items:
        print(color,item)
```

```python
for i in range(1,100,30):
    for j in range (200,300,20):
        print(i*j)
```

```python
# for loop inside while loop
i = 1
while i < 10:
    for j in range(1,10):
        print(i, j)
    i+=1
```

```python
# while loop inside for loop
for i in range(1,10):
    j = 1
    while j < 10:
        print(i, j)
        j+=1
```

```python
i = 0
while i < 5:
    print(i)
    i += 1
    if i == 3:
        break
else:
    print(0)
```

# enumerate function

```python
marks=[5,56,71,89,45,67,90,34,78,98]
index=0
for mark in marks:
    if (index==5):
        print("We are learning pythin ")
    print(mark)
    index +=1
```

```python
# but we can do it with enumerate function more easily
marks=[5,56,71,89,45,67,90,34,78,98]
for index, mark in enumerate (marks):
    if (index==5):
        print("We are learning pythin ")
    print(mark)
```

```python
# but we can do it with enumerate function more easily
marks=[5,56,71,89,45,67,90,34,78,98]
for index, mark in enumerate (marks, start=1):
    if (index==5):
        print("We are learning pythin ")
    print(mark)
```

# lambda function
In Python, a lambda function is a concise way to create anonymous (unnamed) functions. Lambda functions are also known as anonymous functions or lambda expressions. They are defined using the lambda keyword and can take any number of parameters.\
It's important to note that while lambda functions are useful for certain situations, they are limited compared to named functions. Lambda functions are best suited for simple operations, and for more complex functionality, it is often more readable and maintainable to use a regular named function.


```python
def square(number):
    return number*number
print(square(5))
```

```python
def add(a,b):
    return a+b
print(add(2,3))
```

```python
# Factorial
print(7*6*5*4*3*2*1)
print(6*5*4*3*2*1)
print(5*4*3*2*1)
# factorial of 0 is 1 by definition
print(4*3*2*1)
# so we can write that
# factorial (n)=n*factorial(n-1)
```

```python
# Now we can make a function for factorial
def factorial (n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
```

```python
factorial(3)
```

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary with sample data
data = {'Products': ['Product A', 'Product B', 'Product C'],
    'Sales': [100, 150, 80]}

# create a pandas dataframe from the dictionary
df1 = pd.DataFrame(data)

# plot the dataframe as a bar chart
df1.plot(x='Products', y='Sales', kind='bar', legend=False)

# set the y-axis label
plt.ylabel('Sales')

# set the title of the plot
plt.title('Sales of different products')


# display the plot
plt.show()
```

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt
# create a dictionary with sample data
data = {'Products': ['Product A', 'Product B', 'Product C'],
        'Sales': [100, 150, 80]}
# create a pandas dataframe from the dictionary
df1 = pd.DataFrame(data)
# plot the dataframe as a pie chart
df1.set_index('Products')['Sales'].plot(kind='pie', startangle=90)
# set the title of the plot
plt.title('Sales distribution of different products')
# display the plot
plt.ylabel('')  # This is to remove the 'Sales' ylabel which is unnecessary in a pie chart
plt.show()
```

# Histogram

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary with sample data
data = {'Sales': [30,50,120,150,200,300,100, 150, 80, 120, 140, 130, 
                  50,50,30,20,10,50,35,45,55,50,55,65,56,58,58,54,55,50,5,
                  110, 145, 105, 90, 115, 125, 135, 85, 95]}
# create a pandas dataframe from the dictionary
df2 = pd.DataFrame(data)
# plot the dataframe as a histogram
df2['Sales'].plot(kind='hist', bins=5, edgecolor='black')
# set the x-axis and y-axis labels
plt.xlabel('Sales Value')
plt.ylabel('Number of Transactions')
# set the title of the plot
plt.title('Distribution of Sales Transactions')
# display the plot
plt.show()
```

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt

# create a dictionary with sample data
data = {
    'Sales': [100, 150, 80, 120, 140, 130, 110, 145, 105, 90],
    'Profit': [50, 70, 40, 60, 75, 65, 55, 72, 52, 45]
}

# create a pandas dataframe from the dictionary
df3 = pd.DataFrame(data)

# plot the dataframe as a scatter plot
df3.plot(x='Sales', y='Profit', kind='scatter')

# set the x-axis and y-axis labels
plt.xlabel('Sales Value')
plt.ylabel('Profit Value')

# set the title of the plot
plt.title('Relationship between Sales and Profit')

# display the plot
plt.show()
```

```python
# import pandas and matplotlib.pyplot libraries
import pandas as pd
import matplotlib.pyplot as plt
# create a dictionary with sample data
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Monthly Sales': [100, 110, 105, 115, 120, 125, 130, 135, 140, 145, 150, 155]}
# create a pandas dataframe from the dictionary
df4 = pd.DataFrame(data)
# plot the dataframe as a line chart
df4.plot(x='Month', y='Monthly Sales', kind='line', marker='o')
# set the x-axis and y-axis labels
plt.xlabel('Month')
plt.ylabel('Sales Value')
# set the title of the plot
plt.title('Monthly Sales Trend over a Year')
# display the plot
plt.show()
```

```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample correlation matrix
data = {'Product A': [1, 0.6, 0.8, 0.3],
    'Product B': [0.6, 1, 0.5, 0.7],
    'Product C': [0.8, 0.5, 1, 0.4],
    'Product D': [0.3, 0.7, 0.4, 1]}
# Create a pandas dataframe from the dictionary
df5 = pd.DataFrame(data)
# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df5, annot=True, cmap='viridis', vmin=0, vmax=1)
# Set the title of the plot
plt.title('Product Correlation Heatmap')
# Display the plot
plt.show()
```

```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
# Sample data
data = {'Product A': [95, 85, 90, 100, 105, 110],
    'Product B': [100, 105, 95, 90, 115, 120],
    'Product C': [105, 110, 115, 120, 125, 130]}
# Create a pandas dataframe from the dictionary
df6 = pd.DataFrame(data)
# Plotting the boxplot
df6.boxplot(grid=False, vert=True, fontsize=10)
# Set the title of the plot
plt.title('Sales Distribution by Product')
# Set the y-axis label
plt.ylabel('Sales Value')
# Display the plot
plt.show()
```

```python
# we can do the same thing with seaborn library as follows
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample data
data = {'Product A': [95, 85, 90, 100, 105, 110],
    'Product B': [100, 105, 95, 90, 115, 120],
    'Product C': [105, 110, 115, 120, 125, 130]}
# Create a pandas dataframe from the dictionary
df6 = pd.DataFrame(data)
# Plotting the boxplot using seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=df6)
# Set the title of the plot
plt.title('Sales Distribution by Product')
# Set the y-axis label
plt.ylabel('Sales Value')
# Display the plot
plt.show()
```

```python
# import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Sample data
data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],'Sales': [100, 150, 180, 220, 270]}  # Accumulated sales for demonstration}
# Create a pandas dataframe from the dictionary
df7 = pd.DataFrame(data)
# Plotting the area chart
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
plt.fill_between(df7['Month'], df7['Sales'], color="green", alpha=0.4)
plt.plot(df7['Month'], df7['Sales'], color="yellow", alpha=0.6)
# Set the title, x-axis and y-axis labels
plt.title('Monthly Accumulated Sales')
plt.xlabel('Month')
plt.ylabel('Accumulated Sales Value')
# Display the plot
plt.show()
```
