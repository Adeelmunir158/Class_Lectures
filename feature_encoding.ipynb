```python
import sklearn
import pandas as pd
from sklearn.preprocessing import  MinMaxScaler
data={"value": [5, 10, 15, 20, 25]}
df=pd.DataFrame(data)
df.head()
```

```python
#Min mas scalling
scaler = MinMaxScaler()
scaled_df1=scaler.fit(df)
scaled_df1=pd.DataFrame(scaled_df1.transform(df), columns=df.columns)
df["scaled_df1"]=scaled_df1["value"]
df.head()
```

```python
# Standard scaler
from sklearn.preprocessing import  StandardScaler

scaler = StandardScaler()
scaled_df2=scaler.fit(df)
scaled_df2=pd.DataFrame(scaled_df2.transform(df), columns=df.columns)
df["scaled_df2"]=scaled_df2["value"]
df.head()
```

```python
# Robust scaler
from sklearn.preprocessing import  RobustScaler
scaler = RobustScaler()
scaled_df3=scaler.fit(df)
scaled_df3=pd.DataFrame(scaled_df3.transform(df), columns=df.columns)
df["scaled_df3"]=scaled_df3["value"]
df.head()
```

```python
# Lograthmic scaling
import numpy as np
data={"value": [50000, 100000, 150000, 200000, 250000]}
df5=pd.DataFrame(data)
df5.head()
```

```python
df5["log"]=np.log(df5["value"])
df5["log2"]=np.log2(df5["value"])
df5["log10"]=np.log10(df5["value"])
df5.head()
```

```python
# Label encoding
from sklearn.preprocessing import LabelEncoder
data={"Animal": ["cat", "dog", "sheep", "hen", "camel"]}
df7=pd.DataFrame(data)
df7.head()
```

```python
label_encoder=LabelEncoder()
df7["label"]=label_encoder.fit_transform(df7["Animal"])
df7.head()
```

```python
# ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
data={"Size":["Small", "Medium", "Large", "Extra Large"]}
df8=pd.DataFrame(data)
df8.head()
```

```python
ordinal_encoder=OrdinalEncoder()
df8["label"]=ordinal_encoder.fit_transform(df8[["Size"]])
df8.head()
```
